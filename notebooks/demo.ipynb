```python
data = df.loc[1:][['Brand & Content %', 'Marketing Operations %', 'Digital Marketing %', 'Customer Care %', 'Indirect Retail / General Trade %', 'Direct Retail %', 'Direct Sales %', 'Sales Systems Support %', 'Market Share']]
```

```python
# Extract feature (X) and target (y) columns
feature_cols = data.columns[-1]  # all columns but last are features
target_col = list(data.columns[:-1])  # last column is the target/label
# feature_cols = 'Market Share'  # all columns but last are features
# target_col = ['target']  # last column is the target/label
print("Feature column(s):-\n{}".format(feature_cols))
print("Target column: {}".format(target_col))

X_all = data[feature_cols]  # feature values for all students
y_all = data[target_col]  # corresponding targets/labels
print("\nFeature values:-")
print(X_all.head())  # print the first 5 rows
```

```python
# Shuffle the data or randomly select samples to avoid any bias due to ordering in the dataset
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_all, y_all, train_size = 0.8, random_state=1)

print("Training set: {} samples".format(X_train.shape[0]))
print("Test set: {} samples".format(X_test.shape[0]))
```

```python
y_train
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

def model():
    model = Sequential()
    model.add(Dense(600, input_dim=1, kernel_initializer='normal', activation='sigmoid'))
    model.add(Dense(8, activation='sigmoid', kernel_initializer='normal'))
    model.compile(loss='mse',  optimizer='adam')
    return model

estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasRegressor(build_fn=model, epochs=1000, batch_size=10, verbose=1)))
pipeline = Pipeline(estimators)
kfold = KFold(n_splits=5)
results = cross_val_score(pipeline, X_train.values.reshape(-1, 1), y_train, cv=kfold)
print("Model: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```

```python
pipe = Pipeline([
        ('scale', StandardScaler()),
        ('clf', KerasRegressor(model, epochs=1000, batch_size=5, verbose=0))])
pipe.fit(X_train.values.reshape(-1, 1), y_train)
```

# Normalize variables

We are not interested in what could happen if we had infinite amount of money. We are interested in given this much money how best to optimize headcount to maximise market share. Can we assume that market share is effected by number of employees? Can we assume that staffing level attracts market share? This is a hypothesis. 

Assumptions:
* headcounts in each territory remained the same since go live 
* cNPS and eNPS in each territory remained the same since go live
